# Importando Bibliotecas


```python
import pandas as pd 
```

# Importando base de dados

```python
dados = pd.read_csv('Mall_Customers.csv')
```

# Tratamento da base de dados

```python
# Mostrando base de dados
dados
```

```python
# Renomeando base de dados para Português
dados.rename({'CustomerID':'ID_Consumidor','Gender':'Sexo','Age':'Idade','Annual Income (k$)':'Salário Anual (Dollar)','Spending Score (1-100)':'Pontuação Gastos'}, axis= 1, inplace=True)
```

```python
# Renomeando atributos da coluna Sexo
dados['Sexo'].replace(['Male','Female'],['Homem','Mulher'], inplace=True)
dados
```

```python
# Verificação se há dados que faltam
dados.isnull().sum()
```

```python
# Descrição dos dados numéricos
dados.describe()
```

```python
# Descrição dos dados categóricos
dados.describe(include='O')
```

```python
# Média de idade dos consumidores
dados['Idade'].mean()
```

```python
# Média do salário anual dos consumidores
dados['Salário Anual (Dollar)'].mean()
```

```python
# Cliente mais velho da base de dados
dados['Idade'].max()
```

```python
# Cliente mais novo da base de dados
dados['Idade'].min()
```

```python
# Média de vinte consumidores no tocante a sexo, salário anual e pontuação de gastos
dados.groupby(['Sexo','Salário Anual (Dollar)','Pontuação Gastos']).mean().head(20)
```

```python
# Tabela cruzada para saber informações sobre pontuação de gastos versus sexo.
pd.crosstab(dados['Sexo'],dados['Pontuação Gastos'],margins=True).style.background_gradient(cmap='Blues')
```

```python
# Pontuação de gastos,idade,Salário anual de todos os consumidores homens com gastos iguais ou acima de 70
dados[['ID_Consumidor','Salário Anual (Dollar)','Idade','Sexo','Pontuação Gastos']].loc[(dados['Sexo']=='Homem')&(dados['Pontuação Gastos']>=70)]
                                    
```

```python
dados[['ID_Consumidor','Salário Anual (Dollar)','Idade','Sexo','Pontuação Gastos']].loc[(dados['Sexo']=='Mulher')&(dados['Pontuação Gastos']>=70)]
```

```python
# Criando variavel 'Dummy' para transformar dados categóricos em numéricos
pd.get_dummies(dados,columns=['Sexo'])
```

```python
# Média do gastos dos consumidores baseando-se na Pontuação Gastos
dados['Pontuação Gastos'].mean()
```

```python
# Maior indice da idade de um consumidor
dados['Idade'].idxmax()
```

```python
# Menor indice da idade de um consumidor
dados['Idade'].idxmin()
```
